```python
import pandas as pd

df = pd.DataFrame(
    {
        "series_id": ["A", "A", "A", "A", "A", "B", "B", "B", "B", "B"],
        "date": pd.to_datetime(
            [
                "2025-01-01",
                "2025-01-02",
                "2025-01-03",
                "2025-01-04",
                "2025-01-05",
                "2025-01-01",
                "2025-01-02",
                "2025-01-03",
                "2025-01-04",
                "2025-01-05",
            ]
        ),
        "observed_demand": [10, 12, 8, 14, 16, 100, 90, 110, 95, 105],
    }
)

df = df.sort_values(["series_id", "date"]).reset_index(drop=True)
df
```

```python
grouped = df.groupby("series_id", sort=False)
grouped
```

```python
for series_name, group in grouped:
    print(f"\n===== Grupo {series_name} =====")
    display(group)
```

```python
grouped_observed = grouped["observed_demand"]
grouped_observed
```

```python
for series_name, values in grouped_observed:
    print(f"\n===== observed_demand de la serie {series_name} =====")
    print(values)
```

```python
shifted_1 = grouped["observed_demand"].shift(1)
shifted_1
```

```python
df["demand_lag_1"] = grouped["observed_demand"].shift(1)
df
```

```python
df["bad_shift_without_groupby"] = df["observed_demand"].shift(1)
df
```

```python
df["roll_mean_3"] = grouped["observed_demand"].transform(
    lambda series: series.shift(1).rolling(window=3, min_periods=3).mean()
)

df
```

```python
df["roll_sum_3"] = grouped["observed_demand"].transform(
    lambda series: series.shift(1).rolling(window=3, min_periods=3).sum()
)

df["roll_std_3"] = grouped["observed_demand"].transform(
    lambda series: series.shift(1).rolling(window=3, min_periods=3).std()
)

df
```

```python
for series_name, group in df.groupby("series_id", sort=False):
    print(f"\n===== Serie {series_name} =====")
    display(
        group[
            [
                "series_id",
                "date",
                "observed_demand",
                "demand_lag_1",
                "roll_mean_3",
                "roll_sum_3",
                "roll_std_3",
            ]
        ]
    )
```

```python
serie_a = df[df["series_id"] == "A"].copy()

serie_a["shift_1"] = serie_a["observed_demand"].shift(1)
serie_a["manual_roll_mean_3"] = (
    serie_a["shift_1"].rolling(window=3, min_periods=3).mean()
)
serie_a["manual_roll_sum_3"] = serie_a["shift_1"].rolling(window=3, min_periods=3).sum()
serie_a["manual_roll_std_3"] = serie_a["shift_1"].rolling(window=3, min_periods=3).std()

serie_a
```
